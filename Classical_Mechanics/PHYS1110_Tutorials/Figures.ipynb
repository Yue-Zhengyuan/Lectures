```python
import numpy as np
import sys
import os
import glob
from copy import copy
from itertools import product
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
plt.style.use('default')
%config InlineBackend.figure_format = 'svg'
```

```python
fig, ax = plt.subplots(3, 1, figsize=(6,8), sharex=True)
fig.patch.set_facecolor('#FFFFFF')
# adjust spacing
plt.subplots_adjust(left=0.15, bottom=0.15, right=0.95, top=0.9, 
                    wspace=0.3, hspace=0.4)
ax[0].set_xlim(left=0, right=2.5)
for axis in ax:
    axis.xaxis.set_major_locator(MultipleLocator(0.5))
    axis.yaxis.set_major_locator(MultipleLocator(1))
    axis.grid(True, which='major', linestyle='-', lw=1)
    axis.set_ylim(bottom=-2, top=2)
```

```python
for axis in ax:
    axis.clear()
    axis.xaxis.set_major_locator(MultipleLocator(0.5))
    axis.yaxis.set_major_locator(MultipleLocator(1))
    axis.grid(True, which='major', linestyle='-', lw=1)
    axis.set_ylim(bottom=-2, top=2)
t = np.linspace(0, 2.5, 200)
ax[0].plot(t, np.cos(8*np.pi*t), color='blue', label=r'$\cos(8\pi t)$')
ax[0].legend(loc='upper right')
ax[0].set_ylabel(r"$y_1$")

ax[1].plot(t, np.cos(10*np.pi*t), color='blue', label=r'$\cos(10\pi t)$')
ax[1].legend(loc='upper right')
ax[1].set_ylabel(r"$y_2$")

ax[2].plot(t, np.cos(8*np.pi*t) + np.cos(10*np.pi*t), color='blue', label=r'$\cos(8\pi t) + \cos(10 \pi t)$')
ax[2].plot(t, 2 * np.cos(np.pi*t), color='red', linestyle=':', label=r'$\pm 2 \cos(\pi t)$')
ax[2].plot(t, -2 * np.cos(np.pi*t), color='red', linestyle=':')
ax[2].legend(loc='upper right')
ax[2].set_ylabel(r"$y_1 + y_2$")
ax[2].set_xlabel(r"$t$")
fig.tight_layout()
fig
```
